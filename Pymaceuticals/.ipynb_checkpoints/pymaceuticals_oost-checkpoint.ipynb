```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

study_results.head()
```

```python
mouse_metadata.head()
```

```python
# Combine the data into a single DataFrame
df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how='outer')
    

# Display the data table for preview
df.head()
```

```python
#looking at my dataframe
df.head()
```

```python
#Looking at the dataframe part 2
df.info()
```

```python
#counting the mice
mouse_count = df["Mouse ID"].nunique()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
#Mouse g989 was a double
df.groupby(["Mouse ID", "Timepoint"]).size().sort_values()
```

```python
#Located the double mouse and looked at it's data
mask = df["Mouse ID"] == "g989"
df.loc[mask]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

mask = df["Mouse ID"] != "g989"
df2 = df.loc[mask].reset_index(drop=True)

df2.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count2 = df2["Mouse ID"].nunique()
mouse_count2
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
df2.describe()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.

avg_tumor = df2.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
med_tumor = df2.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
var_tumor = df2.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
std_tumor = df2.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
sem_tumor = df2.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()
```

```python
# Assemble the resulting series into a single summary DataFrame.
tumor_df= pd.DataFrame({
    "Average_Tumor": avg_tumor,
    "Median_Tumor": med_tumor,
    "Variance": var_tumor,
    "Standard_Deviation": std_tumor,
    "SEM": sem_tumor
})

tumor_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
#Thanks to Booth for help

final_time = df2.groupby("Mouse ID").Timepoint.max().reset_index()
# Start by getting the last (greatest) timepoint for each mouse

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
df_last = pd.merge(final_time, df2, on=["Mouse ID", "Timepoint"], how="left")

#List of the drugs we need to analyze.
# Capomulin, Ramicane, Infubinol, and Ceftamin

drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

#Create a mask to find the location of these drugs. 
mask = df_last["Drug Regimen"].isin(drugs)

#Update the df
df_last = df_last.loc[mask].reset_index(drop=True)
df_last
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
sns.boxplot(data=df_last, x="Drug Regimen", y="Tumor Volume (mm3)", color="purple")

plt.title("Tumor Volume for Drug Regimen", fontweight="bold", fontsize= 17, fontfamily="Arial")
plt.xlabel("Drug Regimen", fontweight="bold", fontsize= 17, fontfamily="Arial")
plt.ylabel("Tumor Volume (mm3)", fontweight="bold", fontsize= 17, fontfamily="Arial")
plt.grid(axis="y", linestyle="--")
```

```python
# Generate a volin plot that shows the distribution of the tumor volume for each treatment group.
sns.violinplot(data=df_last, x="Drug Regimen", y="Tumor Volume (mm3)", color="lightblue")

plt.title("Tumor Volume for Drug Regimen", fontweight="bold", fontsize= 17, fontfamily="Arial")
plt.xlabel("Drug Regimen", fontweight="bold", fontsize= 17, fontfamily="Arial")
plt.ylabel("Tumor Volume (mm3)", fontweight="bold", fontsize= 17, fontfamily="Arial")
plt.grid(axis="y", linestyle="--")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
# Add the linear regression equation and line to plot
x_values = df_cap["Weight (g)"]
y_values = df_cap["Tumor Volume (mm3)"]

# do regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept # these are the points/predictions
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
```

```python
#Ripped from 5.3.7
# Change Canvas Size
plt.figure(figsize=(12, 6))

# Create a basic plot
plt.scatter(df_cap["Weight (g)"], df_cap["Tumor Volume (mm3)"], color="purple")
plt.plot(df_cap["Weight (g)"], regress_values, color="black", linewidth=4) # the regression line

# Customizations

# Change colors
# Add in reference lines
# Add Labels/Title
plt.xlabel("Weight (g)", fontfamily="Arial", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontfamily="Arial", fontsize=12)
plt.title("Tumor Volume vs. Weight in Capomulin", fontweight="bold", fontsize=16, fontfamily="Arial")

# Set X/Y Limits
#plt.ylim(10, 16)
plt.xlim(10, 30)

# Add in a background grid
plt.grid(linestyle="--", color="lightgrey", alpha=0.75)

# Annotate
plt.annotate(line_eq, (24, 25), fontsize=15, color="black")

# Show/Save the Graph
plt.show()
```
